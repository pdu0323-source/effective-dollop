```python
import random, math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import hsv_to_rgb
import ipywidgets as widgets
from ipywidgets import interact
import pandas as pd
import os
from IPython.display import display

# --- CSV Palette Management (Assuming this structure is desired) ---

# Global variable for the palette CSV file name
PALETTE_FILE = "palette.csv"

# Initialize palette.csv if not exists
if not os.path.exists(PALETTE_FILE):
    df_init = pd.DataFrame([
        {"name":"sky", "r":0.4, "g":0.7, "b":1.0},
        {"name":"sun", "r":1.0, "g":0.8, "b":0.2},
        {"name":"forest", "r":0.2, "g":0.6, "b":0.3}
    ])
    df_init.to_csv(PALETTE_FILE, index=False)

def read_palette():
    return pd.read_csv(PALETTE_FILE)

def add_color(name, r, g, b):
    df = read_palette()
    df = pd.concat([df, pd.DataFrame([{"name":name,"r":r,"g":g,"b":b}])], ignore_index=True)
    df.to_csv(PALETTE_FILE, index=False)
    print(f"Added {name}")

def update_color(name, r=None, g=None, b=None):
    df = read_palette()
    if name in df["name"].values:
        idx = df.index[df["name"]==name][0]
        if r is not None: df.at[idx,"r"] = r
        if g is not None: df.at[idx,"g"] = g
        if b is not None: df.at[idx,"b"] = b
        df.to_csv(PALETTE_FILE, index=False)
        print(f"Updated {name}")
    else:
        print(f"{name} not found")

def delete_color(name):
    df = read_palette()
    df = df[df["name"]!=name]
    df.to_csv(PALETTE_FILE, index=False)
    print(f"Deleted {name}")

def load_csv_palette():
    df = read_palette()
    return [(row.r, row.g, row.b) for row in df.itertuples()]

palette_csv = load_csv_palette()
print("Loaded palette:", palette_csv)

# Show CSV palette (assuming necessary for the structure)
def show_palette(palette):
    plt.figure(figsize=(6,2))
    for i, c in enumerate(palette):
        plt.fill_between([i, i+1], 0, 1, color=c)
        plt.text(i+0.5, -0.1, f"{i+1}", ha="center", va="top")
    plt.axis("off")
    plt.show()

show_palette(palette_csv)

# --- Core Drawing Functions (Modified Variables) ---

# Blob shape
def blob(center=(0.5, 0.5), r=0.3, points=200, wobble=0.15):
    """Generate a wobbly closed shape around a center with base radius r."""
    angles = np.linspace(0, 2*math.pi, points, endpoint=False)
    radii  = r * (1 + wobble*(np.random.rand(points)-0.5))
    x = center[0] + radii * np.cos(angles)
    y = center[1] + radii * np.sin(angles)
    return x, y

def make_palette(k=8, mode="pastel", base_h=0.60): # k를 8로 늘림
    """Return k colors (RGB) by sampling HSV depending on mode."""
    cols = []
    if mode == "csv":
        return load_csv_palette()

    for _ in range(k):
        if mode == "pastel":
            h = random.random(); s = random.uniform(0.1,0.3); v = random.uniform(0.9,1.0)
        elif mode == "vivid":
            h = random.random(); s = random.uniform(0.9,1.0);  v = random.uniform(0.9,1.0) # 더 강렬하게 수정
        elif mode == "mono":
            h = base_h;         s = random.uniform(0.4,0.7);   v = random.uniform(0.6,1.0) # 모노톤 채도 높임
        else: # random
            h = random.random(); s = random.uniform(0.4,1.0); v = random.uniform(0.7,1.0)
        cols.append(tuple(hsv_to_rgb([h,s,v])))
    return cols

# Main drawing function (Added save_file parameter)
def draw_poster(n_layers=15, wobble=0.25, palette_mode="vivid", seed=123, save_file=""):
    """Draws the abstract poster based on UI parameters, saves if save_file is provided."""

    # Check if a figure already exists from a previous interact call
    # This is necessary to properly control figure lifecycle in ipywidgets
    try:
        plt.close(plt.gcf())
    except:
        pass

    random.seed(seed); np.random.seed(seed)
    # 캔버스 크기를 가로형으로 수정
    fig, ax = plt.subplots(figsize=(9,6))
    ax.axis('off')
    # 배경색을 연한 청색 계열로 수정
    ax.set_facecolor((0.90, 0.95, 1.0))

    # 팔레트 크기 8개로 생성
    palette = make_palette(8, mode=palette_mode)

    # 레이어 루프
    for _ in range(n_layers):
        cx, cy = random.random(), random.random()
        # 반지름 범위 수정
        rr = random.uniform(0.18, 0.50)
        x, y = blob((cx,cy), r=rr, wobble=wobble)
        color = random.choice(palette)
        # 투명도 범위를 더 부드럽게 수정
        alpha = random.uniform(0.2, 0.5)
        ax.fill(x, y, color=color, alpha=alpha, edgecolor=(0,0,0,0))

    # 제목 레이블 (폰트 크기 수정)
    ax.text(0.05, 0.95, f"Interactive Poster • {palette_mode}",
            transform=ax.transAxes, fontsize=14, weight="bold")

    # PDF 저장 로직 추가
    save_path = save_file.strip()
    if save_path and save_path.endswith(".pdf"):
        try:
            # dpi=300은 고해상도 출력을 위한 설정입니다.
            fig.savefig(save_path, dpi=300, bbox_inches="tight", pad_inches=0.1)
            print(f"\n포스터가 성공적으로 저장되었습니다: {save_path}")
        except Exception as e:
            print(f"\n저장 중 오류 발생: {e}")
    elif save_path:
         print(f"\n경고: 저장하려면 파일 이름이 '.pdf'로 끝나야 합니다. (예: my_poster.pdf)")

    plt.show()

# --- Custom Widget Setup for Robust Access ---

# 1. 위젯 객체 생성 (이것을 명시적으로 참조하여 값 가져오기)
n_layers_w = widgets.IntSlider(min=5,max=25,step=1,value=15, description="Layers (레이어)")
wobble_w = widgets.FloatSlider(min=0.05,max=0.5,step=0.01,value=0.25, description="Wobble (흔들림)")
palette_w = widgets.Dropdown(options=["pastel","vivid","mono","random","csv"], value="vivid", description="Palette (팔레트)")
seed_w = widgets.IntSlider(min=0,max=9999,step=1,value=123, description="Seed (시드)")

# 2. PDF 저장 UI 구성
pdf_filename_input = widgets.Text(
    value="my_abstract_poster.pdf",
    description="파일명:",
    placeholder="my_poster.pdf",
    style={'description_width': 'initial'}
)
save_button = widgets.Button(description="PDF로 저장 실행")
output_area = widgets.Output()

def on_save_button_clicked(b):
    """버튼 클릭 시 draw_poster 함수를 호출하여 저장합니다."""
    with output_area:
        output_area.clear_output()
        # 3. 명시적으로 생성된 위젯 객체의 .value를 사용하여 값 가져오기 (가장 안정적인 방법)
        draw_poster(
            n_layers=n_layers_w.value,
            wobble=wobble_w.value,
            palette_mode=palette_w.value,
            seed=seed_w.value,
            save_file=pdf_filename_input.value
        )

# 버튼 클릭 이벤트 연결
save_button.on_click(on_save_button_clicked)

# 4. 인터랙트 위젯 구성 (위젯 객체를 직접 전달)
# interact는 이 함수를 실행하고 위젯을 표시하며, 변경 시마다 draw_poster를 호출합니다.
interact_output = interact(
    draw_poster,
    n_layers=n_layers_w,
    wobble=wobble_w,
    palette_mode=palette_w,
    seed=seed_w,
    save_file=widgets.fixed("") # interact에서는 저장 함수를 숨깁니다.
)

# 5. PDF 저장 UI를 인터랙티브 UI 아래에 배치
display(
    widgets.VBox([
        widgets.HTML("<h3>PDF 저장 설정</h3>"),
        widgets.HBox([pdf_filename_input, save_button]),
        output_area
    ])
)
```
