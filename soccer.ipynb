```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random

# =========================================================
# 1. í™•ì¥ëœ ë°ì´í„° ì •ì˜ (Data: ì¶•êµ¬íŒ€ ìƒ‰ìƒ ì •ë³´)
# (íŒ€ ì´ë¦„, ë©”ì¸ ìƒ‰ìƒ HEX, ë³´ì¡° ìƒ‰ìƒ HEX, íŒ¨í„´/ë°°ê²½ ìƒ‰ìƒ HEX)
# HEX ì½”ë“œëŠ” ê³µì‹ í™ˆí˜ì´ì§€, ìœ ë‹ˆí¼, ì— ë¸”ëŸ¼ ë“±ì„ ì°¸ê³ í•˜ì—¬ ì„¤ì •ë˜ì—ˆìŠµë‹ˆë‹¤.
# =========================================================

TEAM_COLORS = {
    # í”„ë¦¬ë¯¸ì–´ ë¦¬ê·¸ (2024/25 ì‹œì¦Œ ê¸°ì¤€ ì£¼ìš” íŒ€ ë° ëª¨ë“  íŒ€ í¬í•¨ ê°€ì •)
    "Arsenal FC": ["#EF0107", "#00366E", "#9C824A"],
    "Aston Villa": ["#4F0024", "#95BFE5", "#67A6CC"],
    "Chelsea FC": ["#030948", "#FFFFFF", "#DA291C"],
    "Liverpool FC": ["#C8102E", "#00A389", "#FFFFFF"],
    "Manchester City": ["#6CABDD", "#1C2C5B", "#FFFFFF"],
    "Manchester United": ["#DA291C", "#FBE122", "#000000"],
    "Tottenham Hotspur": ["#FFFFFF", "#132257", "#CE1126"],
    "Newcastle United": ["#000000", "#FFFFFF", "#5BA8A6"],
    "Everton FC": ["#003399", "#FFFFFF", "#418A37"],
    "West Ham United": ["#7A263A", "#F3D45F", "#1BB190"],
    "Leicester City": ["#003090", "#FDBE11", "#FFFFFF"],
    # ... (ìƒëµëœ ë‹¤ë¥¸ EPL íŒ€ë“¤ì„ ì—¬ê¸°ì— ì¶”ê°€í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤) ...

    # ë¼ ë¦¬ê°€
    "Real Madrid": ["#FFFFFF", "#005697", "#000000"],
    "FC Barcelona": ["#A50044", "#004D98", "#FDBE11"],
    "Atletico Madrid": ["#CB3524", "#FFFFFF", "#272D30"],
    "Real Betis": ["#008653", "#FFFFFF", "#964B00"],

    # ë¶„ë°ìŠ¤ë¦¬ê°€
    "Bayern Munich": ["#DC052D", "#FFFFFF", "#0066CC"],
    "Borussia Dortmund": ["#FDE100", "#000000", "#FFFFFF"],
    "Bayer Leverkusen": ["#E3221C", "#000000", "#FFFFFF"],
    "RB Leipzig": ["#001C53", "#FFD700", "#FFFFFF"],

    # ë¦¬ê·¸ 1 (í”„ë‘ìŠ¤)
    "Paris Saint-Germain": ["#004A95", "#DA291C", "#FFFFFF"],
    "AS Monaco": ["#C5AA74", "#FFFFFF", "#E30613"],
    "Lille OSC (Lille)": ["#004481", "#E63E3F", "#FFFFFF"],
    "Olympique de Marseille (ë§ˆë¥´ì„¸ìœ )": ["#00468C", "#FFFFFF", "#000000"],

    # ì„¸ë¦¬ì— A (ì´íƒˆë¦¬ì•„)
    "Juventus": ["#FFFFFF", "#000000", "#999999"],
    "Inter Milan (ì¸í…Œë¥´)": ["#004A95", "#FFFFFF", "#000000"],
    "AC Milan": ["#FF0000", "#000000", "#FFFFFF"],
    "AS Roma (ASë¡œë§ˆ)": ["#86273B", "#F5B41E", "#FFFFFF"],
    "Napoli": ["#007AC9", "#FFFFFF", "#123062"],

    # Kë¦¬ê·¸
    "FC Seoul": ["#86242B", "#FFFFFF", "#000000"],
}

# =========================================================
# 2. ê·¸ë¦¼ ìƒì„± í•¨ìˆ˜ (Generative Art Function)
# ì½”ë“œëŠ” ì´ì „ê³¼ ë™ì¼í•˜ê²Œ ìœ ì§€ë˜ì–´ ë°ì´í„°ë§Œ ë³€ê²½ë¨
# =========================================================

def generate_abstract_art(colors, seed=None):
    """
    ì£¼ì–´ì§„ ìƒ‰ìƒ ë°°ì—´ì„ ì‚¬ìš©í•˜ì—¬ ì¶”ìƒì ì¸ ì  íŒ¨í„´ ì´ë¯¸ì§€ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
    """
    if seed is not None:
        random.seed(seed)
        np.random.seed(seed)

    fig, ax = plt.subplots(figsize=(6, 6), facecolor='none')
    ax.set_aspect('equal')
    ax.set_facecolor(colors[2]) # ë°°ê²½ìƒ‰ì€ íŒ¨í„´ ìƒ‰ìƒìœ¼ë¡œ ì„¤ì •

    num_points = 500
    # íŒ¨í„´ì´ ì¤‘ì•™ì— ì§‘ì¤‘ë˜ë„ë¡ ì •ê·œ ë¶„í¬ ì‚¬ìš©
    x = np.random.normal(0, 1, num_points)
    y = np.random.normal(0, 1, num_points)

    # ì  í¬ê¸°ë¥¼ ëœë¤í•˜ê²Œ ì„¤ì •
    sizes = np.random.uniform(20, 300, num_points)

    # ìƒ‰ìƒ ëª©ë¡ì—ì„œ ëœë¤í•˜ê²Œ ìƒ‰ìƒ ì„ íƒ (ë©”ì¸/ë³´ì¡° ìƒ‰ìƒë§Œ ì‚¬ìš©)
    point_colors = random.choices(colors[:2], k=num_points)

    # ì  íŒ¨í„´ ìƒì„±
    scatter = ax.scatter(
        x, y,
        s=sizes,
        c=point_colors,
        alpha=0.6,
        edgecolors=colors[2], # ê²½ê³„ì„ ì€ íŒ¨í„´ ìƒ‰ìƒ
        linewidths=1
    )

    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.axis('off') # ì¶• ì œê±°
    plt.tight_layout()

    return fig

# =========================================================
# 3. Streamlit ì•± êµ¬ì„±
# =========================================================

def main():
    st.set_page_config(page_title="ì¶•êµ¬íŒ€ ìƒì„± ì˜ˆìˆ ", layout="wide")
    st.title("âš½ ì¶•êµ¬íŒ€ ìƒ‰ìƒì„ í™œìš©í•œ ìƒì„± ì˜ˆìˆ  í”„ë¡œì íŠ¸")
    st.markdown("ë°ì´í„° ê¸°ë°˜ ë””ìì¸ê³¼ ìƒì„± ì˜ˆìˆ ì˜ ê²°í•©")
    st.markdown("---")

    # ì‚¬ì´ë“œë°” ì„¤ì •
    st.sidebar.header("ğŸ¨ ì„¤ì •")

    # ëª¨ë“  íŒ€ ëª©ë¡ì„ ì •ë ¬í•˜ì—¬ ì…€ë ‰íŠ¸ ë°•ìŠ¤ì— í‘œì‹œ
    team_list = sorted(TEAM_COLORS.keys())
    selected_team = st.sidebar.selectbox(
        "íŒ€ì„ ì„ íƒí•˜ì„¸ìš”:",
        team_list
    )

    # ëœë¤ ì‹œë“œ ì…ë ¥ (ì¬í˜„ì„±ì„ ìœ„í•´)
    seed = st.sidebar.number_input(
        "ê·¸ë¦¼ Seed ì…ë ¥ (ìˆ«ìë¥¼ ë°”ê¾¸ë©´ ë‹¤ë¥¸ ê·¸ë¦¼ì´ ìƒì„±ë©ë‹ˆë‹¤)",
        min_value=1,
        max_value=10000,
        value=random.randint(1, 10000)
    )

    # íŒ€ ë¶„ë¥˜ ì •ë³´ í‘œì‹œ
    if selected_team in ["Arsenal FC", "Aston Villa", "Chelsea FC", "Liverpool FC", "Manchester City", "Manchester United", "Tottenham Hotspur", "Newcastle United", "Everton FC", "West Ham United", "Leicester City"]:
        league = "ğŸ´ó §ó ¢ó ¥ó ®ó §ó ¿ í”„ë¦¬ë¯¸ì–´ ë¦¬ê·¸ (Premier League)"
    elif selected_team in ["Real Madrid", "FC Barcelona", "Atletico Madrid", "Real Betis"]:
        league = "ğŸ‡ªğŸ‡¸ ë¼ ë¦¬ê°€ (La Liga)"
    elif selected_team in ["Bayern Munich", "Borussia Dortmund", "Bayer Leverkusen", "RB Leipzig"]:
        league = "ğŸ‡©ğŸ‡ª ë¶„ë°ìŠ¤ë¦¬ê°€ (Bundesliga)"
    elif selected_team in ["Paris Saint-Germain", "AS Monaco", "Lille OSC (Lille)", "Olympique de Marseille (ë§ˆë¥´ì„¸ìœ )"]:
        league = "ğŸ‡«ğŸ‡· ë¦¬ê·¸ 1 (Ligue 1)"
    elif selected_team in ["Juventus", "Inter Milan (ì¸í…Œë¥´)", "AC Milan", "AS Roma (ASë¡œë§ˆ)", "Napoli"]:
        league = "ğŸ‡®ğŸ‡¹ ì„¸ë¦¬ì— A (Serie A)"
    elif selected_team == "FC Seoul":
        league = "ğŸ‡°ğŸ‡· Kë¦¬ê·¸ 1 (K League 1)"
    else:
        league = "ê¸°íƒ€ ë¦¬ê·¸"


    if selected_team:
        st.subheader(f"ì„ íƒ íŒ€: {selected_team} ({league})")

        # 1. ìƒ‰ìƒ ë°ì´í„° ì¶”ì¶œ
        colors_data = TEAM_COLORS[selected_team]

        st.subheader("ì‚¬ìš©ëœ íŒ€ ëŒ€í‘œ ìƒ‰ìƒ (HEX ì½”ë“œ)")
        col1, col2, col3 = st.columns(3)

        with col1:
            st.markdown(f"**ë©”ì¸ ìƒ‰ìƒ:** `{colors_data[0]}`")
            st.markdown(f"<div style='background-color:{colors_data[0]}; height:30px; border-radius: 5px; border: 1px solid #ccc;'></div>", unsafe_allow_html=True)

        with col2:
            st.markdown(f"**ë³´ì¡° ìƒ‰ìƒ:** `{colors_data[1]}`")
            st.markdown(f"<div style='background-color:{colors_data[1]}; height:30px; border-radius: 5px; border: 1px solid #ccc;'></div>", unsafe_allow_html=True)

        with col3:
            st.markdown(f"**íŒ¨í„´/ë°°ê²½ ìƒ‰ìƒ:** `{colors_data[2]}`")
            st.markdown(f"<div style='background-color:{colors_data[2]}; height:30px; border-radius: 5px; border: 1px solid #ccc;'></div>", unsafe_allow_html=True)

        st.markdown("---")

        # 2. ì´ë¯¸ì§€ ìƒì„± ë° í‘œì‹œ
        st.subheader("ìƒì„±ëœ ì¶”ìƒí™” íŒ¨í„´ (Generative Art Output)")

        # ìƒì„± í•¨ìˆ˜ í˜¸ì¶œ
        fig = generate_abstract_art(colors_data, seed=seed)

        # Streamlitì— Matplotlib ê·¸ë¦¼ í‘œì‹œ
        st.pyplot(fig)

        # ê·¸ë¦¼ ë‹¤ìš´ë¡œë“œ ë²„íŠ¼
        st.markdown("**---**")
        col_dl, col_blank = st.columns([1, 4])

        with col_dl:
            # ë‹¤ìš´ë¡œë“œ íŒŒì¼ ì´ë¦„ ì„¤ì •
            file_name = f"{selected_team.replace(' ', '_')}_art_{seed}.png"

            # Matplotlib ìº”ë²„ìŠ¤ë¥¼ PNG ë°ì´í„°ë¡œ ë³€í™˜
            buf = fig.get_figure().canvas.buffer_rgba()

            st.download_button(
                label="ğŸ–¼ï¸ ì´ë¯¸ì§€ ë‹¤ìš´ë¡œë“œ (PNG)",
                data=buf.tobytes(),
                file_name=file_name,
                mime="image/png"
            )
        st.markdown("---")
        st.info("ì´ ì½”ë“œëŠ” íŒ€ ìƒ‰ìƒ ë°ì´í„°ì™€ ëœë¤ ì‹œë“œë¥¼ ê¸°ë°˜ìœ¼ë¡œ ì ë“¤ì˜ ìœ„ì¹˜ì™€ í¬ê¸°ë¥¼ ë¬´ì‘ìœ„ë¡œ ìƒì„±í•˜ì—¬ ì¶”ìƒì ì¸ íŒ¨í„´ì„ ë§Œë“­ë‹ˆë‹¤. 'Seed' ê°’ì„ ë³€ê²½í•˜ì—¬ ë‹¤ë¥¸ íŒ¨í„´ì„ ìƒì„±í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")


# ì•± ì‹¤í–‰ (Colab í™˜ê²½ì—ì„œ í…ŒìŠ¤íŠ¸ ì‹œ ì•„ë˜ ì½”ë“œë¥¼ ë³„ë„ ì…€ì— ì…ë ¥í•˜ê³  ì‹¤í–‰)
if __name__ == "__main__":
    main()
```
